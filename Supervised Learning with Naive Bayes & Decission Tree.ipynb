# 2. Naive Bayes

```python
df.drop(["PassengerId",	"Name" ,"SibSp", "Ticket", "Fare", "Embarked", "Parch"], axis = "columns", inplace = True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies = pd.get_dummies(inputs.Sex).astype(int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

Saya menghapus kolom'male' karena kita hanya butuh satu kolom untuk menyatakan 'male' atau 'female'.

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# 3. DecissionTree

```python
# Read file csv
df1 = pd.read_csv("C:/Dokumen/1. SAINS DATA/SEMESTER 4/Machine Learning/RTM 3/train.csv")

# Menampilkan beberapa baris pertama pada dataset
df1.head(3)
```

```python
df1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df1.head(5)
```

```python
#drop missing value
df1.dropna(inplace=True)
df1.info()
```

```python
df1[~df1.duplicated()]
#dinegasikan agar yang keluar itu yg false yaitu yg bukan duplicate
df1.reset_index(inplace=True)
df1.info()
```

```python
df1.drop('index',axis='columns',inplace=True)
df1.head(2)
```

```python
inputs = df1.drop('Survived',axis='columns')
target = df1.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
dummies = pd.get_dummies(inputs.Sex).astype(int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import tree

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=1)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Hitung akurasi model
accuracy = model.score(X_test, y_test)
print("\nAkurasi model:", accuracy)

# Visualisasikan pohon keputusan dengan kedalaman maksimum
plt.figure(figsize=(13, 6))
tree.plot_tree(model, feature_names=inputs.columns, class_names=['0', '1'], filled=True, max_depth=2)
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import tree

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=1)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Hitung akurasi model
accuracy = model.score(X_test, y_test)
print("\nAkurasi model:", accuracy)

# Visualisasikan pohon keputusan dengan kedalaman maksimum
plt.figure(figsize=(13, 13))
tree.plot_tree(model, feature_names=inputs.columns, class_names=['0', '1'], filled=True, max_depth=20)
plt.show()
```
